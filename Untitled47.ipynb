```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/Raotbl6.csv')
df
```

```python
df['date']=pd.to_datetime(df['date'])
df=df.set_index(['date'])
```

```python
from statsmodels.tsa.statespace.varmax import VARMAX
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import grangercausalitytests,adfuller
```

```python
plt.figure(figsize=(10,8))
plt.subplot(4,2,1)
plt.plot(df['rgnp'])
plt.xlabel('rgnp')
plt.title('RGNP')
plt.subplot(4,2,2)
plt.plot(df['pgnp'])
plt.xlabel('pgnp')
plt.title('PGNP')
plt.subplot(4,2,3)
plt.plot(df['ulc'])
plt.xlabel('ulc')
plt.title('ULC')
```

```python
dftest=adfuller(df['rgnp'])
print('rgnp')
print('1.ADF STATISTICS:',dftest[0])
print('2.P-Value:',dftest[1])
```

```python
if dftest[1]>=0.05:
  print('Not Stationary')
else:
  print('Stationary')
```

```python
#Grangercausalitytests
print('rgnp causes  ulc?\n')
print('------')
granger_1=grangercausalitytests(df[['ulc','rgnp']],4)
print('ulc causes rgnp?\n')
print('--------')
granger_2=grangercausalitytests(df[['rgnp','ulc']],4)
```

```python
macro_data=df[['rgnp','ulc']]
macro_data
```

```python
macro_data.shape
```

```python
train_df=macro_data[:-12]
test_df=macro_data[-12:]
```

```python
model=VAR(train_df.diff()[1:])
sorted_order=model.select_order(maxlags=20)
print(sorted_order.summary())
```

```python
var_model=VARMAX(train_df,order=(4,0),enforce_stationarity=True)
fitted_model=var_model.fit(disp=False)
print(fitted_model.summary())
```

```python
n_forecast=12
predict=fitted_model.get_prediction(start=len(train_df),end=len(train_df)+n_forecast-1)
predictions=predict.predicted_mean
predictions.columns=['ulc_predicted','rgnp_predicted']
predictions
```

```python
test_vs_pred=pd.concat([test_df,predictions],axis=1)
test_vs_pred.plot(figsize=(10,7))
```
